```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, desc
```

```python
spark = SparkSession.builder.appName('sparksql').getOrCreate()
```

```python
print(spark.version)
```

```python
data = spark.read.format('csv')\
.option('inferSchema','true')\
.option('header','true')\
.option('path','file:///home/krish/Spark_practice/operations_management.csv')\
.load()
```

```python
data
```

```python
data.printSchema()
```

```python
data_2 = data.select('industry','value').filter((col('value')>200)&(col('industry')!='total')).orderBy(desc('value'))
```

```python
data_2.printSchema()
```

```python
data_2.show(5)
```

```python
data_2.createOrReplaceTempView('data')
```

```python
spark.sql("""select industry,value 
from data 
where value>200 
and industry !="total"
""").show(5)
```
